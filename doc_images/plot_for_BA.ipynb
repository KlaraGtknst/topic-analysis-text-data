# Infer time

```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
times = [899.218, 159.627]
settings = ['Precomputed (GloVe)', 'Custom (Word2Vec)']
title = 'InferSent time per embedding'


fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(2)
y_pos = np.arange(len(settings))

bars = ax.barh(y_pos, times, align='center')
ax.bar_label(bars, label_type='center')
ax.set_yticks(y_pos, labels=settings)
#ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Time (seconds)')
ax.set_title(title)

#plt.savefig(re.sub(' ', '_', title) + '.pdf', format="pdf", bbox_inches='tight')
plt.show()
```

```python
import statistics

inverse_emb = np.array([0,2])
embedding = np.array([0,1])
cos_sim = statistics.mean([np.dot(inverse_emb, embedding) /(np.linalg.norm(inverse_emb)*np.linalg.norm(embedding))])

print(cos_sim)
```

# Time per embeddings

```python
all_times = {'InferSent + AE': 4366.564, 'SBERT': 51.135, 'TFIDF + AE': 17.086, 'doc2vec': 6.890, 'USE': 5.600}
```

```python
title = 'Time necessary to compute 2048 embeddings'


fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(5)
y_pos = np.arange(len(all_times.keys()))

bars = ax.barh(y_pos, all_times.values(), align='center')
ax.bar_label(bars, label_type='edge')
ax.set_yticks(y_pos, labels=all_times.keys())
ax.set_xlim(0, 5000)
#ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Time (seconds)')
ax.set_title(title)

#plt.savefig(re.sub(' ', '_', title) + '.pdf', format="pdf", bbox_inches='tight')
plt.show()
```
