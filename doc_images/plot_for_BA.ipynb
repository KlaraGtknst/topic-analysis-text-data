# Infer time

```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
times = [899.218, 159.627]
settings = ['Precomputed (GloVe)', 'Custom (Word2Vec)']
title = 'InferSent time per embedding'


fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(2)
y_pos = np.arange(len(settings))

bars = ax.barh(y_pos, times, align='center')
ax.bar_label(bars, label_type='center')
ax.set_yticks(y_pos, labels=settings)
#ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Time (seconds)')
ax.set_title(title)

#plt.savefig(re.sub(' ', '_', title) + '.pdf', format="pdf", bbox_inches='tight')
plt.show()
```
